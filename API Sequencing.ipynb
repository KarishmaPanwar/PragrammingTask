```python
numerical_columns = ['inter_api_access_duration(sec)', 'api_access_uniqueness', 'sequence_length(count)', 
                     'vsession_duration(min)', 'num_sessions', 'num_users', 'num_unique_apis']
for col in numerical_columns:
    threshold = supervised_dataset[col].mean()  
    supervised_dataset[col] = (supervised_dataset[col] > threshold).astype(int)
supervised_dataset['ip_type'] = supervised_dataset['ip_type'].apply(lambda x: 0 if x == 'default' else 1)
supervised_dataset['source'] = supervised_dataset['source'].apply(lambda x: 0 if x == 'E' else 1)
supervised_dataset['classification'] = supervised_dataset['classification'].apply(lambda x: 0 if x == 'normal' else 1)
supervised_dataset.drop('_id', axis=1, inplace=True)
# supervised_dataset.drop('Unnamed: 0', axis=1, inplace=True)
supervised_dataset
```

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

min_support = 0.2  
frequent_itemsets = apriori(supervised_dataset, min_support=min_support, use_colnames=True)
```

```python
min_confidence = 0.5  
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=min_confidence)
```

```python
print("Frequent Itemsets:")
print(frequent_itemsets)
print("\nAssociation Rules:")
print(rules)
```

```python
# Apply FP-Growth Algorithm

from mlxtend.frequent_patterns import fpgrowth
min_support = 0.2
frequent_itemsets_fp = fpgrowth(supervised_dataset, min_support=min_support, use_colnames=True)

# Step 2: Generate Association Rules

from mlxtend.frequent_patterns import association_rules

# Set minimum confidence threshold
min_confidence = 0.5

# Generate association rules
rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=min_confidence)

# Display frequent itemsets and association rules
print("Frequent Itemsets:")
print(frequent_itemsets_fp)
print("\nAssociation Rules:")
print(rules_fp)
```
